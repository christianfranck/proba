# Distribution Uniforme

```python
from numpy import *
from matplotlib.pyplot import *

def plot_uniform(a,b): 
    x=linspace(-2,5,10000) 
    y=(1/(b-a))*(sign(x-a)-sign(x-b))/2
    plot(x,y,label="a="+str(a)+" b="+str(b)) 

plot_uniform(0,1)
plot_uniform(2,4)

legend()
show ()
```

# Distribution Normale

```python
from numpy import *
from matplotlib.pyplot import *

def plot_normal(mu,sigma): 
    x=linspace(-5,8,10000) 
    y=exp(-0.5*((x-mu)/sigma)**2)/(sigma*sqrt(2*pi))
    plot(x,y,label="mu="+str(mu)+" sigma="+str(sigma)) 

plot_normal(0,1)
plot_normal(1,2)

legend()
show ()
```

# Distribution Bta

```python
from numpy import *
from matplotlib.pyplot import *

def plot_beta(alpha,beta): 
    x=linspace(0,1,10000) 
    y=(x**(alpha-1)) * ((1-x)**(beta-1)) 
    y=y*10000/sum(y)
    plot(x,y,label="beta("+str(alpha)+","+str(beta)+")") 
    
plot_beta(1,1)
plot_beta(1,3)
plot_beta(5,2)
plot_beta(3,7)

legend()
show ()
```
