# Distribution Geom√©trique

$$P_K(k)=(1-p)^{k-1}\cdot p$$

```python
from math import *
from numpy import *
from matplotlib.pyplot import *

def plot_geometric(p,k_max=10): 
    x=range(0,k_max+1) 
    y=[(p)*((1-p)**(k-1)) for k in x]
    bar(x,y)
    title("$p$="+str(p))
    xlabel("$k\in \{1,2,3,...\}$")
    ylabel("$P_K(k)$")

plot_geometric(0.6)

show()
```

# Distribution Binomiale

$$P_K(k)={n \choose k}\cdot p^k \cdot (1-p)^{n-k}$$

```python
from numpy import *
from matplotlib.pyplot import *

def plot_binomial(p,n): 
    x=range(0,n+1) 
    y=[(factorial(n)/(factorial(k)*factorial(n-k)))*(p**k)*((1-p)**(n-k)) for k in x]
    bar(x,y)
    title("$p$="+str(p)+" $n$="+str(n))
    xlabel("$k$")
    ylabel("$P_K(k)$")

plot_binomial(0.6,6)

show()
```

# Distribution Hyperg√©om√©trique

$$P_K(k)=\frac{{pA \choose k} \cdot {qA \choose {n-k}}} {{pA+qA} \choose n} $$

```python
from numpy import *
from matplotlib.pyplot import *

def choose(k,n):
    return factorial(n)/(factorial(k)*factorial(n-k))

def plot_hypergeometrique(n,pA,qA): 
    x=range(0,n+1) 
    y=[(choose(k,pA)*choose(n-k,qA)/choose(n,pA+qA)) for k in x]
    bar(x,y)
    title("$n$="+str(n)+" "+"$pA$="+str(pA)+" $qA$="+str(qA))
    xlabel("$k$")
    ylabel("$P_K(k)$")

plot_hypergeometrique(6,12,10)

show ()
```

# Loi de Poisson

$$P_K(k)=\frac{\lambda^k}{k!}\cdot e^{-\lambda}$$

```python
from numpy import *
from matplotlib.pyplot import *

def choose(k,n):
    return factorial(n)/(factorial(k)*factorial(n-k))

def plot_poisson(l,n_max=10): 
    x=range(0,n_max+1) 
    y=[((l**k)*exp(-l)/factorial(k)) for k in x]
    bar(x,y)
    title("$\lambda$="+str(l))
    xlabel("$k\in \{1,2,3,...\}$")
    ylabel("$P_K(k)$")

plot_poisson(4)

show ()
```
